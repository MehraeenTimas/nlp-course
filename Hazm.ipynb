1-just a raw persian text sample we're gonna work with

```python
texts = [
    "تهران امروز میزبان نشست اقتصادی بود.",
    "علی رضا با شرکت سامسونگ قرارداد جدیدی امضا کرد.",
    "دانشگاه تهران در سال ۱۳۱۳ تأسیس شد."
]
```

2-installing hazm. we're gonna rely on hazm for this project

```python
!pip install gensim python-crfsuite fasttext flashtext && pip install hazm --no-deps
```

3-removing spaces, half spaces etc...normalizing the text basically

```python
from hazm import Normalizer

normalizer = Normalizer()
texts_norm = [normalizer.normalize(t) for t in texts]
texts_norm
```

4-tokenizing our text and splittin it into words

```python
from hazm import word_tokenize

tokens_list = [word_tokenize(t) for t in texts_norm]
tokens_list
```

5-filtering stopwords

```python
from hazm import stopwords_list

filtered_texts = []

for tokens in tokens_list:
    filtered = [t for t in tokens if t not in stopwords_list()]
    filtered_texts.append(filtered)

print(filtered_texts)
```

6-pos tagging: labeling each word

```python
from hazm import POSTagger
tagger = POSTagger(model='/content/pos_tagger.model')
```

```python
for tokens in filtered_texts:
    pos_tags = tagger.tag(tokens)
    print(pos_tags[:10])
```

7-Name entity recognition

```python
from hazm import *
ner = HazmNER(model_path='ner/model-best')
ner.predict_entity(pos-tags)
ner.predict(pos-tags)
```
